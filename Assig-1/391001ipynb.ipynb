```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/train.csv')
df.head()
```

```python
class TitanicEDA:
    def __init__(self, filepath):
        self.filepath = filepath
        self.data = None

    def load_data(self):
        self.data = pd.read_csv(self.filepath)
        print("Data loaded successfully.")
        return self.data

    def summary_statistics(self):
        if self.data is not None:
            return self.data.describe(include='all')
        else:
            raise ValueError("Data not loaded. Please load the data first.")

    def visualize_survival_rate(self, feature):
        if self.data is None:
            raise ValueError("Data not loaded. Please load the data first.")

        plt.figure(figsize=(10, 6))
        sns.set(style="whitegrid")
        sns.barplot(x=feature, y='Survived', data=self.data)
        plt.title(f'Survival Rate by {feature}', fontsize=15)
        plt.ylabel('Survival Rate')
        plt.xlabel(feature)
        plt.savefig(f"Visualization_{feature}.")
        plt.show()
```

```python
filepath="/content/train.csv"
data_load=TitanicEDA(filepath)
data_load.load_data()
```

```python
data_load.summary_statistics()
```

```python
data_load.visualize_survival_rate("Pclass")
```

```python
data_load.visualize_survival_rate("Sex")
```

```python
data_load.data['AgeBin'] = pd.cut(data_load.data['Age'], bins=[0, 12, 18, 25, 35, 60, 80], labels=['0-12', '13-18', '19-25', '26-35', '36-60', '61-80'])
data_load.visualize_survival_rate('AgeBin')
```
